#### Database 'studio_gross' - Visualization, Analysis and Recommendation

```python
#cleaning by viewing and summing studios by total gross
df_studio.groupby(['studio']).sum()
```

```python
#groupin
df_histo = df_studio.groupby(['studio']).agg('sum')
```

```python
#histogram
df_histo= df_histo.sort_values('total_gross', ascending=False).head(10)
```

```python
#looking at the index
df_histo.index
```

```python
#preparing histo
font = {'family' : 'Helvetica',
        'weight' : 'bold',
        'size'   : 10}

plt.rc('font', **font)
```

```python
#plotting
plt.figure(figsize=(10,8))

studios = df_histo.index
dom_gross = df_histo.domestic_gross
for_gross = df_histo.foreign_gross
tot_gross = df_histo.total_gross

plt.bar(range(len(studios)), dom_gross, color='blue')
plt.bar(range(len(studios)), for_gross, color='#f2ee79', bottom=dom_gross)
plt.title('Top 5 Market Share Big Players ', fontsize=15)
plt.xlabel('Studio Names', fontsize=25)
plt.ylabel('Revenues (in Billions)', fontsize=15)
plt.xticks(range(len(studios)), studios)

plt.legend(['Domestic', 'Foreign'])
plt.show();
```

# 2nd Data Analysis Process: Uses 2 DB's- Join for analysis

### database: df_budget. Containing information about movies, production budgets, domestic and worlwide gross.

```python
#Looking inside the database budget again 
df_budget.head()
```

```python
#looking at info
df_budget.info()
```

```python
#looking at types
df_budget.dtypes
```

```python
#dropping columns
df_budget.drop(['domestic_gross'], axis=1, inplace=True)
```

```python
#finding out data types
df_budget.dtypes
```

```python
# I want to change the columns to plain number integers I can work later with.
df_budget.head(2)
```

```python
# removing dollar signs and commas from dollar amounts
# converting dollar amounts from strings into integers
df_budget['production_budget'] = df_budget['production_budget'].str.replace(',', '').str.replace('$', '').astype(int)
df_budget['worldwide_gross']   = df_budget['worldwide_gross'].str.replace(',', '').str.replace('$', '').astype(int)
df_budget.head(2)
```

### database: df1. Opus. Containg movie titles by genre, budget, and production years- This is available on the web. Downloaded to my pc and brought to for the purpose of this analysis.

```python
#I downloaded a new data base that has genres only joined it with my budget data
df1= pd.read_csv('/Users/jonax/Documents/opus_df.csv')
```

```python
#looking at the head
df1.head()
```

```python
#it's a very clearn data base, I'll change the 'title', to 'movie' to conduct the join
df1.info()
```

```python
# renaming certain columns
df1 = df1.rename(columns={'title':'movie'})
```

```python
#verifying it column has been changed
df1.head(2)
```

```python
#setting the index to movie on this database
df1.set_index('movie', inplace=True)
```

## Conducting an Inner Join :  df_super

```python
#joining two database (inner) at movie columun 
df_super= df_budget.join(df1, how='inner')
df_super
```

#### The  New Joined Dabase characteristics

```python
#infor of new database
df_super.info()
```

```python
#dropping unnecesary columns on my joint database
df_super.drop(['Unnamed: 0', 'release_date', 'creative_type', 'dom_gross', 'prod_method'], axis=1, inplace=True)
```

```python
#looking at the cleaned database
df_super
```

```python
#I'll make the numbers a bit more manageable for both production budget and world gross
df_super ['production_budget']= df_super['production_budget']/1000000
```

```python
df_super ['worldwide_gross']= df_super['worldwide_gross']/1000000
```

```python
#checking it out
df_super
```

### Now that I have a Joined Database (df_super) I'll visualize and analyse to propose recommendations

#### Plotting my chart

```python
# plotting the number of movies per genre in dataset to find out insight and analyse it.
plt.figure(figsize=(12,8))
sns.countplot(x='genre', data=df_super, palette='colorblind')
plt.title('Amount of Movies per Genre', fontsize=18)
plt.ylabel('Count', fontsize=20)
plt.xlabel('Genre', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(rotation=-80);
```

## Recommendiation 2

```python
#For my visualizations, I found it helpful to show my plots both with and without outliers. While outliers paint the whole picture of the data being analyzed, it can be helpful to see the data more closesly and "zoom in" (both literally and figuratively) on the trends without outliers.
# generating box plot of world gross statistics per genre
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
sns.boxplot(x='genre', y='worldwide_gross', data=df_super, palette='dark')
plt.xticks(rotation=-80)
plt.ylabel('World Gross in Millions', fontsize=16)
plt.xlabel('Genre', fontsize = 16)
plt.title('World Gross by Genre', fontsize = 18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14);
#saved in images as fig2
#plt.subplots_adjust(bottom=0.2)
#plt.savefig('./images/fig2.png')
```
